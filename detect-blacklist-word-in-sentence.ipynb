## **IS THERE ANY BLACKLIST WORD IN THE SENTENCE ?**

Given the list of blacklist word, this code is to check whether the sentence that we have input contain the blacklist word or not.

```python
blacklist_word = ["andi","susi","bola"]

#input the sentence
sentence = input("Enter the sentence here :  ")

#turn case of blacklist_word and sentence into lowercase,because if we've input the sentence in different case with blacklist_word, the code will be read as false
blacklist_word = list(map(lambda x: x.lower(), blacklist_word))
sentence = sentence.lower()

#clear space in both tip of the list object in blacklist_word, because if we change it to "  andi" and "andi  ", then if we input "andi makan nasi padang" to the sentence, the code will be read as false
blacklist_word = [x.strip(' ') for x in blacklist_word]

#create a variable as a string to store the value of iteration later on
check = ""  

#using for to iterate over a sequence of list object in blecklist_word and check whether the sentence that we've input contain that list object using if else 
for i in blacklist_word :
  if i in sentence :
    check = "ada blacklist word"
    break
  else :
    check = "tidak ada blacklist word"
  
print(check)
```
