# **IS THE WORD PALINDROME?**

This code is to check whether the word or number that we've input is **palindrome** or not

"A palindrome is a word, phrase, sentence, or even number that reads the same backward and forward."

Example : madam, racecar, level, 123321

```python
#input word or number 
word = input("Enter Your Word Here : ")

#convert all letter of the word into lowercase, because if we are input in different cased "Level is not the same as leveL" the code will be read as false
word = word.lower()

#clear all space in both tip of the word, because if we type "  level" and "level  " as an input, the code will be read as false
word = word.strip()

#create a reverse function using def
def reverse(word) :
  return word [::-1]

#using if else to declare the word or number that have been input is palindrome or not based on certain condition 
if word == reverse(word) :
  print(word + " is a palindrome")
else :
  print(word + " is not a palindrome")
```
