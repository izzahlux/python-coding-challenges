## **Is It A Prime Number Or Not?**

A prime number is a natural number greater than 1, which is only divisible by 1 and itself.

1. prime number is positive integer greater than 1 (x > 1, x = integer and not a decimal number)
2. prime number has no positive divisors other than 1 and itself (create a loop to find the other divisors other than 1 and the number that we've input using for and modulo)
3. example : if we are input x = 4, then we must divide it with number between range(2, 4) = (2, 3). check it one by one, 4 % 2 == 0 is true, so 4 is not a prime number because we've found the other divisors other than 1 and 4, that is 2. if we are input x = 3, then we must divide it with number between range(2, 3) = (2) and 3 % 2 == 0 is false, so 3 is a prime number because we have not found the other divisors other than 1 and 3. 

Limitation :
- If you input a positive integer number, don't input it as decimal number. for example if you want input 13 don't input it as 13.0.

```python
def prime_check(x) : 
  if type(x) == float :
    print(x,"is not a prime number")
  elif type(x) == int :
    if x > 1 :  
      for i in range(2, x) :
        if x % i == 0 :
          print(x,"is not a prime number")
          break
      else :
        print(x,"is a prime number")  
    else :
      print(x,"is not a prime number")
```

```python
prime_check(1)
```

```python
prime_check(5)
```

```python
prime_check(-5)
```

```python
prime_check(8.9)
```
