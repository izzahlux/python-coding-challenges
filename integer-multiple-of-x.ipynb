## **Print Out Integers Which Are Multiples Of 3 From 1 To 22**

1. Create a loop using for x in range (1, 23). We will iterate value of x that are from 1 to 22 into certain condition.
2. We can get the number which are multiple of 3 by using the condition with modulo operation. Any value of x that divide into 3 and the remainder of dividing equal to zero will be print out.

```python
for x in range (1, 23) :
  if x % 3 == 0 :
    print(x)
```
